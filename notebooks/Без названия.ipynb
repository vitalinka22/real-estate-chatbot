```python
#load the data
import pandas as pd

df = pd.read_csv("../data/raw/immo_data.csv")

print(f"Total rows: {len(df)}")
print(f"Total columns {len(df.columns)}")
```

```python
#filter Berlin only 

berlin = df[df["regio2"] == "Berlin"]
print(f"Berlin listings: {len(berlin)}")
```

```python
berlin.isnull().sum()
```

```python
for col in df.columns.tolist():
    print (col)
```

```python
columns_to_keep = [
    'scoutId',           # unique ID
    'regio2', 'regio3',  # city, district
    'street', 'houseNumber', 'geo_plz',  # address
    'totalRent', 'baseRent', 'serviceCharge',  # price
    'livingSpace', 'noRooms', 'floor', 'numberOfFloors',  # size
    'balcony', 'cellar', 'hasKitchen', 'garden', 'lift', 'petsAllowed',  # features
    'typeOfFlat', 'heatingType', 'condition', 'interiorQual',  # type & quality
    'yearConstructed', 'lastRefurbish', 'newlyConst',  # age
    'energyEfficiencyClass', 'noParkSpaces'  # extras
]

berlin_clean = berlin[columns_to_keep].copy()
print(berlin_clean.shape)
```

```python
berlin_clean = berlin_clean.dropna(subset = ['totalRent', 'livingSpace', 'noRooms'])
print(f"After dropping missing price/size : {len(berlin_clean)} rows" )
```

```python
berlin_clean = berlin_clean[
    (berlin_clean['totalRent'] > 200) &
    (berlin_clean['totalRent'] < 10000)
    ]

print(f'After removing outliers: {len(berlin_clean)} rows')
print(berlin_clean['totalRent'].describe())
```

```python
berlin_clean.to_csv('../data/processed/berlin_clean.csv', index = False)
print('Saved!')
```
